```python
#dropped the NaN row
aflights = aflights.drop(aflights.index[2])
aflights
```

```python
#replaced the "NaN" values with the appropriate Airline
aflights['Airline'].fillna(method='ffill', inplace = True)
aflights
```

```python
#used melt function to make the wanted columns into variables, all of my minutes turned into one column
longflights = aflights.melt(id_vars = ['Airline', 'Status'],  value_vars = ['Los Angeles', 'Phoenix', 'San Diego', 'San Fran', 'Seattle'])
longflights
```

```python
#minor adjustment to have columns with relevant names
longflights = longflights.rename(columns={"variable": "City", "value": "Frequency"})
longflights
```

Data is now tidy.

```python
#making an Alaska only DF
Alaska = (longflights.loc[longflights['Airline'] == 'ALASKA'])
Alaska
```

```python
#number of times Alaska airport was on time regardless of city
(Alaska[Alaska['Status'] == 'on time']['Frequency'].sum())
```

Alaska had 3,274 on time flights, regardless of city.

```python
#ratio of Alaska's on time flights versus all flights
(Alaska[Alaska['Status'] == 'on time']['Frequency'].sum())/(Alaska['Frequency'].sum())
```

Out of all of Alaska's flights in our dataset, Alaska is on time roughly 87% of the time.

```python
#making an AmWest only DF
AmWest = (longflights.loc[longflights['Airline'] == 'AM WEST'])
AmWest
```

```python
#AmWest's total flights that were on time.
(AmWest[AmWest['Status'] == 'on time']['Frequency'].sum())
```

AmWest had 6,438 on time flights, regardless of city.

```python
#AmWest's ratio of on time flights over total flights.
(AmWest[AmWest['Status'] == 'on time']['Frequency'].sum())/(AmWest['Frequency'].sum())
```

Out of all of Alaska's flights in our dataset, AmWest is on time roughly 89% of the time.
This proves to be a better ratio, even given AmWest's greater quantity of flight demand. AmWest is a more punctual airline in both capacities.
